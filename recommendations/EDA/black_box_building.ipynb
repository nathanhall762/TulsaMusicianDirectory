## Grab data given generated by Ethan from the TMD Directory

```python
## grab data from file
with open("../data/audioFeaturesFromTMD.json", 'r') as file:
    # Load the JSON data from the file
    data = json.load(file)
```

## Convert the data into matching format that comes from Brayden's Cloud Function.
### This needs to eventually be packaged into a module for future use as the database grows.


```python
TMDartistAudioFeatures = {}
artistAverageAudioFeatures = {}
for artist in data:
    TMDartistAudioFeatures[artist['artistId']] = []
    for album in artist['albumTracksFeatures']:
        TMDartistAudioFeatures[artist['artistId']].append(album['trackFeatures'])

for key, value in TMDartistAudioFeatures.items():
    # print(key, value)
    sum_values = {feature: 0.0 for feature in value[0][0].keys()}
    song_count = len(value)
    # print(song_count)
    for song in value[0]:
        for feature, value in song.items():
            if feature == 'id':
                continue
            sum_values[feature] += value
    del sum_values['id']
    artistAverageAudioFeatures[key] = {feature: value/song_count for feature, value in sum_values.items()}


pp(artistAverageAudioFeatures)
```

## Store the new formatted data that will be hosted with the module for Cosine Sim, ED, etc.

```python
with open('../data/artistAverageAudioFeatures.json', 'w') as file:
    json.dump(artistAverageAudioFeatures, file, indent=4)
```

### Get user average for all songs (equally weighted sadly)

```python
def getUserAverage(songList) -> dict:
    """Gets general values from the song values from the Playlist"""
    dictValues = sorted(songList[0].keys())
    if 'speechiness' in dictValues:
        dictValues.remove('speechiness')
    
    sum_values = {feature: 0.0 for feature in dictValues}
    song_count = len(songList)
    for song in songList:
        for feature, value in song.items():
            if feature in dictValues:
                sum_values[feature] += value
    
    valueDict = {feature: value/song_count for feature, value in sum_values.items()}
    return valueDict
```
