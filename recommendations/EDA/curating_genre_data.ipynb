```python
import pandas as pd
from pprint import pprint as pp
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
```

```python
genre_playlists = pd.read_csv('../data/playlists_per_category.csv', index_col=0)
genre_playlists
genres_to_keep = genre_playlists.loc[['Top Lists',
 'Hip-Hop',
 'Pop',
 'Country',
 'Latin',
 'Rock',
 'R&B',
 'Dance/Electronic',
 'Indie',
 'Jazz',
 'Folk & Acoustic',
 'Ambient',
 'Alternative',
 'Classical',
 'Soul',
 'Punk',
 'Blues',
 'Afro',
 'Metal',
 'Funk & Disco']]
genres_to_keep
```

## Make dictionary of {genres: playlists} for parsing

```python
genres_dict = genres_to_keep.to_dict(orient='index')

for genre in genres_dict:
    p_list = [value for value in genres_dict[genre].values() if not pd.isna(value)]
    genres_dict[genre] = p_list
```

## Pull song ids for each playlist

```python
playlists_songs_df = pd.read_csv('../data/playlists_songs.csv', index_col=0)
playlists_songs_df
```

### get list of song ids for each playlist into a dictionary {playlist_id: [song_ids]}

```python
playlists_songs_dict = playlists_songs_df.to_dict(orient='index')

for playlist in playlists_songs_dict:
    songs_list = [value for value in playlists_songs_dict[playlist].values() if not pd.isna(value)]
    set(songs_list)
    playlists_songs_dict[playlist] = list(songs_list)
```

### Parse through dictionary and fill the song ids for each playlist

```python
genre_to_songIds_dict = {}
for genre in tqdm(genres_dict):
    print(genre)
    print(genres_dict[genre])
    print('------------------')
    song_ids = []
    for playlist in genres_dict[genre]:
        song_ids.extend(playlists_songs_dict[playlist])
    genre_to_songIds_dict[genre] = song_ids
```

## now make a dictionary that stores all of the audio features for each song

```python
song_features_df = pd.read_csv('../data/songs_audio_features.csv', index_col=0)
song_features_df
```

```python
final_dict = {}

for genre in tqdm(genre_to_songIds_dict):
    
    try:
        print(genre)
        sum_values = {feature: 0.0 for feature in song_features_df.columns}
        song_count = len(genre_to_songIds_dict[genre])
        for song in genre_to_songIds_dict[genre]:
            try:
                song_af = song_features_df.loc[song]
                # print(song_af)
                # print(type(song_af))
                for feature, value in song_af.items():
                    if feature in sum_values:
                        sum_values[feature] += value
            except:
                print('song not found')
                song_count -= 1
                continue

        final_dict[genre] = {feature: value/song_count for feature, value in sum_values.items()}
    except:
        print('error')
        continue

pp(final_dict)

# print(song_features_df.loc["5eVxIY8kTGe9czhM5SCeR0"].to_json())
```

```python
final_dict_df = pd.DataFrame.from_dict(final_dict, orient='index')
final_dict_df.sort_index(axis=1, inplace=True)
final_dict_df.to_json('../data/genre_features.json', orient='index')
final_dict_df
```
