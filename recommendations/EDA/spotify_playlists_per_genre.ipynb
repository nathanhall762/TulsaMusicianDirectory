```python

playlists = {}

# Get a list of popular playlists
def get_popular_playlists(limit=50):
    playlists = sp.featured_playlists(country='US', limit=limit)
    return playlists['playlists']['items']


# Print the names of popular playlists

popular_playlists = get_popular_playlists(50)
for playlist in popular_playlists:
    playlists[playlist['id']] = playlist['name']

pprint(playlists)

```

```python
# Get category playlists from all US categories


def get_category_playlists(country='US', limit=50):

    category_list = {}
    offset = 0

    while True:
        categories = sp.categories(country=country, limit=limit, offset=offset)
        for category in categories['categories']['items']:
            category_list[category['id']] = category['name']
        if categories['categories']['next'] is not None:
            offset += limit
        else:
            break

    return category_list

us_categories = get_category_playlists()

print(f'Number of categories: {len(us_categories)}')
pprint(us_categories)
```
